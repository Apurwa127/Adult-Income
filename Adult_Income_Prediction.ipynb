```python
import pandas as pd
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import Lasso, LinearRegression

def hyper_par_tuning(x, y):
    model_params = {
        'Linear_Regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'Lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [0.1, 1.0, 10.0],
                'fit_intercept': [True, False]
            }
        },
        'Decision_Tree': {
            'model': DecisionTreeClassifier(),
            'params': {
                'criterion': ['gini', 'entropy'],
                'splitter': ['best', 'random']
            }
        }
    }

    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    models = []

    for name, config in model_params.items():
        clf = GridSearchCV(config['model'], config['params'], cv=cv)
        clf.fit(x, y)
        models.append({
            'model_name': name,
            'model_score': clf.best_score_,
            'model_params': clf.best_params_
        })

    return pd.DataFrame(models, columns=['model_name', 'model_score', 'model_params']).sort_values(by = 'model_score', ascending  = False)

hyper_par_tuning(x, y)
```

```python
best_param = {
    'criterion': 'gini',
    'splitter': 'best'
    }
dt  = DecisionTreeClassifier(**best_param)
dt.fit(x,y)
```

```python
import numpy as np

def predict_salary(country, X, occupation, age, workclass, education, educationnum, race, gender, capitalgain, capitalloss, hoursperweek):
    find_column = np.where(X.columns == f"occupation_{occupation}")[0]
    x = np.zeros(len(X.columns))
    x[0] = country
    x[1] = age
    x[2] = workclass
    x[3] = education
    x[4] = educationnum
    x[5] = race
    x[6] = gender
    x[7] = capitalgain
    x[8] = capitalloss
    x[9] = hoursperweek
    if len(find_column) > 0:
        x[find_column[0]] = 1

    return dt.predict([x])[0]
```
